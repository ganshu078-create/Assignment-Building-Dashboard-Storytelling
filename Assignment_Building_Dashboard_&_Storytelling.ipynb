# Q1. Explain the difference between a Tableau Dashboard and a Tableau Story Point. Give two healthcare-related use cases.

## Answer

Tableau Dashboard and Tableau Story are both used to present data insights, but they serve different purposes.

### Tableau Dashboard

A Dashboard is a collection of multiple visualizations displayed on a single screen. It allows users to monitor and interact with data through filters, charts, and KPIs.

### Tableau Story Point

A Story Point is a sequence of visualizations that presents data in a narrative format. It helps explain insights step-by-step and guides users through a logical flow of analysis.


## Difference Between Dashboard and Story Point

| Feature | Dashboard | Story Point |
|----------|-----------|-------------|
| Purpose | Monitor data in one place | Present insights as a story |
| Structure | Multiple charts on one screen | Sequence of visualizations |
| User Experience | Interactive analysis | Guided presentation |
| Usage | Daily monitoring | Business presentations |
| Focus | Current performance | Data storytelling |


## Healthcare Use Cases

### Use Case 1: Hospital Performance Dashboard

A hospital administrator can monitor:
- Total Patients
- Treatment Cost
- Department Performance
- Patient Outcomes

using a single interactive dashboard.

### Use Case 2: Patient Care Story

Management can create a story showing:
- Patient Distribution
- Department Performance
- Treatment Trends
- Recovery Outcomes

to explain healthcare performance over time.

## Conclusion

Dashboards are ideal for monitoring KPIs, while Story Points are useful for presenting insights through a structured narrative.


# Q2. What are Actions in Tableau? Explain how Filter Action can help hospital management analyze patient data.

## Answer

Actions in Tableau allow users to interact with visualizations and dynamically explore data.

Common Tableau Actions include:

- Filter Action
- Highlight Action
- URL Action

### What is a Filter Action?

A Filter Action allows users to click on a chart element and automatically filter related visualizations.

### Healthcare Example

Suppose a dashboard contains:

- Region Chart
- Department Chart
- Patient Outcome Chart

When management clicks on the "North Region," all other charts automatically update to display data only for that region.

### Benefits

- Faster analysis
- Interactive reporting
- Better decision-making
- Easy data exploration

## Conclusion

Filter Actions improve dashboard interactivity and help hospital management focus on specific patient groups, departments, or regions.


# Q3. Patient Distribution Analysis

## Objective

Analyze the total number of patients across hospital regions.

### Visualization Used

Bar Chart

### Steps

1. Drag Hospital Region to Rows.
2. Drag Patient ID to Columns.
3. Change aggregation to COUNT.
4. Add labels.
5. Sort in descending order.

### Fields Used

| Shelf | Field |
|---------|--------|
| Rows | Hospital Region |
| Columns | COUNT(Patient ID) |

### Story Point Title

```text
Patient Distribution Across Hospital Regions
```

### Insight

The chart identifies regions with the highest and lowest patient volumes.

### Business Importance

Helps management allocate staff, beds, and medical resources efficiently.

## Conclusion

Regional patient analysis supports better operational planning and resource allocation.
